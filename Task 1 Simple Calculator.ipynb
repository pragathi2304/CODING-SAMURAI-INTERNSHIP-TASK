# Simple Calculator Project

##  Description:
This GUI calculator is built using Python’s Tkinter library as part of the **Coding Samurai Internship (Batch A-49)**.

##  Features:
- Basic operations: +, –, ×, ÷
- Clear and Exit buttons
- GUI layout using `Tkinter`

##  Instructions:
To run this project, click **Run All** or run the code cell. A new window will open with the calculator interface.


```python
import tkinter as tk

# Function to update the display with user input
def update_display(value):
    current = display_var.get()
    if current == "0":
        display_var.set(value)
    else:
        display_var.set(current + value)

# Function to clear the display
def clear_display():
    display_var.set("0")

# Function to evaluate the expression and show result
def calculate_result():
    try:
        result = eval(display_var.get())
        display_var.set(str(result))
    except Exception:
        display_var.set("Error")

# Create main window
root = tk.Tk()
root.title("Tkinter Calculator")
root.configure(bg="lightgray")
root.resizable(False, False)

# StringVar to hold and update the display value
display_var = tk.StringVar(value="0")

# Display label (acts as the screen of the calculator)
display_label = tk.Label(
    root, textvariable=display_var,
    font=("Arial", 24), bg="lightgray", anchor="e",
    padx=10, pady=10, relief="ridge", borderwidth=3
)
display_label.grid(row=0, column=0, columnspan=4, padx=5, pady=10, sticky="we")

# Button configuration
buttons = [
    ("7", 1, 0), ("8", 1, 1), ("9", 1, 2), ("/", 1, 3),
    ("4", 2, 0), ("5", 2, 1), ("6", 2, 2), ("*", 2, 3),
    ("1", 3, 0), ("2", 3, 1), ("3", 3, 2), ("-", 3, 3),
    ("0", 4, 0), (".", 4, 1), ("=", 4, 2), ("+", 4, 3),
]

# Create and place buttons
for (text, row, col) in buttons:
    if text == "=":
        command = calculate_result
    else:
        command = lambda t=text: update_display(t)
    
    tk.Button(
        root, text=text, command=command,
        width=5, height=2, font=("Arial", 18),
        bg="white"
    ).grid(row=row, column=col, padx=5, pady=5)

# Clear button
tk.Button(
    root, text="C", command=clear_display,
    width=11, height=2, font=("Arial", 18),
    bg="#f2a2a2"
).grid(row=5, column=0, columnspan=2, padx=5, pady=10)

# Exit button
tk.Button(
    root, text="Exit", command=root.destroy,
    width=11, height=2, font=("Arial", 18),
    bg="#ccc"
).grid(row=5, column=2, columnspan=2, padx=5, pady=10)

# Start the GUI event loop
root.mainloop()
```
